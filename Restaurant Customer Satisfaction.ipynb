#### SCALING

```python
df_num_unscaled = df_num.copy()
df_num_scaled = df_num.copy()
```

```python
df_num_scaled.drop(['OnlineReservation','DeliveryOrder','LoyaltyProgramMember'],axis=1,inplace=True)
df_num_scaled
```

```python
df_num_unscaled.drop(['Age', 'Income', 'AverageSpend', 'GroupSize','WaitTime', 'ServiceRating',
       'FoodRating', 'AmbianceRating'],axis=1,inplace=True)
df_num_unscaled
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler,LabelEncoder
mm = MinMaxScaler()
df_num_scaled = pd.DataFrame(mm.fit_transform(df_num_scaled),columns = df_num_scaled.columns)
df_num_scaled
```

```python
df_num = pd.concat([df_num_scaled,df_num_unscaled],axis = 1)
```

```python
df_num
```

#### ENCODING

```python
for i in df_cat.columns:
    print(i)
    print(df_cat[i].unique())
```

```python
df_cat['Gender'].replace({'Male':1,'Female':0},inplace=True)
df_cat['VisitFrequency'].replace({'Rarely':0,'Daily':1,'Weekly':2,'Monthly':3},inplace=True)
df_cat['PreferredCuisine'].replace({'American':0,'Chinese':1,'Indian':2,'Italian':3,'Mexican':4},inplace=True)
df_cat['TimeOfVisit'].replace({'Breakfast':0,'Lunch':1,'Dinner':2},inplace=True)
df_cat['DiningOccasion'].replace({'Business':0,'Casual':1,'Celebration':2},inplace=True)
df_cat['MealType'].replace({'Takeaway':0,'Dine-in':1},inplace=True)
```

```python
for i in df_cat.columns:
    print(i)
    print(df_cat[i].unique())
```

#### MERGING

```python
df_final = pd.concat([df_num,df_cat],axis = 1)
df_final
```

#### TRAIN TEST SPLIT

```python
x = df_final
y = df['HighSatisfaction']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.3, random_state = 15)
xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

###### =========================================================================================================

#### MODELLING

##### Logistic Model (BASE MODEL)

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import f1_score,classification_report,confusion_matrix,accuracy_score
lr = LogisticRegression()
lr_model = lr.fit(xtrain,ytrain)

lr_train_pred = lr_model.predict(xtrain)
lr_test_pred = lr_model.predict(xtest)

print('Training Accuracy:',lr.score(xtrain,ytrain)*100)
print('Testing Accuracy:',lr.score(xtest,ytest)*100)
```

```python
print('Training Classfication Report:\n',classification_report(ytrain,lr_train_pred))
```

```python
print('Testing Classfication Report:\n',classification_report(ytest,lr_test_pred))
```

#### Cross Val Score (Logistic Regression)

```python
from sklearn.model_selection import cross_val_score
lr_cvs = cross_val_score(lr,x,y,cv = 10)
lr_cvs
```

```python
sns.boxplot(lr_cvs)
```

##### <font color='Orange'>INFERENCE :</font>

    * The scores range from 0.8467 to 0.94, indicating that the logistic regression model performs consistently well across different folds.
    * The relatively small variation in scores suggests that the model is stable and not highly sensitive to the particular training or validation sets used.

##### Decision Tree Model 

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 15)
dt_model = dt.fit(xtrain,ytrain)

dt_train_pred = dt_model.predict(xtrain)
dt_test_pred = dt_model.predict(xtest)

print('Training Accuracy:',dt.score(xtrain,ytrain)*100)
print('Testing Accuracy:',dt.score(xtest,ytest)*100)
```

##### Decision Tree Model  (HYPER PARAMETER - TUNING)

```python
%%time
from sklearn.model_selection import RandomizedSearchCV
params = [{'criterion':['gini','entropy'],
           'max_depth':range(1,30),
           'min_samples_split':range(3,30),
           'min_samples_leaf':range(1,30),}]

grid = RandomizedSearchCV(estimator = DecisionTreeClassifier(random_state = 15),param_distributions = params, 
                          cv = 20)
grid_model = grid.fit(xtrain,ytrain)

best_param = grid_model.best_params_
print('Best Parameters:',best_param)
```

##### Decision Tree Tunned Model 

```python
dt_tunned = DecisionTreeClassifier(**best_param,random_state = 15)
dt_tunned_model = dt_tunned.fit(xtrain,ytrain)

dt_tunned_train_pred = dt_tunned_model.predict(xtrain)
dt_tunned_test_pred = dt_tunned_model.predict(xtest)

print('Training Accuracy:',dt_tunned.score(xtrain,ytrain)*100)
print('Testing Accuracy:',dt_tunned.score(xtest,ytest)*100)
```
